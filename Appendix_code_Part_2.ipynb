```python
import pandas as pd
from statsmodels.formula.api import ols
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('cleaned_data.csv', index_col = 0)
# --------------------------------------------------------
df1['log_Quantity'] = np.log(df1["Quantity"])
df1['log_Price'] = np.log(df1["Price"])
df1['sqrt_Quantity'] = np.sqrt(df1["Quantity"])
df1['sqrt_Price'] = np.sqrt(df1["Price"])
# df = df[df['Price']<650]
# df = df[df['Gap']<340]
df1 = df1.drop(df1[(df1['Gap']>360)].index)
df1 = df1.drop(df1[(df1['Price']>300) & (df1['Quantity']>125)].index) 
df1 = df1.drop(df1[(df1['Price']>1000)].index)
df1 = df1.drop(df1[(df1['Price']>750) & (df1['isNormCabin']==0)].index) 
df1 = df1.drop(df1[(df1['Price']>400) & (df1['isNormCabin']==1)].index) 
df1 = df1.drop(df1[(df1['Price']>650) & (df1['Customer_Cat']=='B')].index) 
df1 = df1.drop(df1[(df1['Price']>700) & (df1['Customer_Cat']=='A')].index)
df = df1
print(len(df))
```

```python
df.head()
```

## Original Q - Modeling
### Strutual form ols

```python
# stucture form ols
# use structural form to do linear regression
structural_form = ols('Quantity  ~ Price + Train_Number_All + isNormCabin + isWeekend + Customer_Cat + Gap + isdate_1819', df).fit()
structural_form.summary2()
```

### 2SLS

```python
# 2SLS 
# reduced form regression
Price_reduced_form = ols('Price ~ Train_Number_All + isNormCabin + isWeekend + Customer_Cat + Gap + isdate_1819 + isReturn + isOneway', df).fit()
print(Price_reduced_form.summary())
# get seat_price predictions
df["pre_Price"] = Price_reduced_form.predict(df[['Train_Number_All','isNormCabin','isWeekend','isdate_1819', 'Customer_Cat','Gap','isReturn','isOneway']])
# 2nd step regression
struc_2sls_form = ols('Quantity  ~  pre_Price + Train_Number_All + isNormCabin + isWeekend + Customer_Cat + Gap + isdate_1819', df).fit()
print(struc_2sls_form.summary2())
```

### Test of endogeneity

```python
# Test of endogeneity
df['res1stage'] = Price_reduced_form.resid
hausman_model = ols('Quantity  ~ Price + Train_Number_All + isNormCabin + isWeekend + Customer_Cat + Gap + isdate_1819 + res1stage', df).fit()
hausman_model.summary2()
```

### Test for overidentification restrictions

```python
# test for overidentification restrictions
predicted_quantity = struc_2sls_form.predict(df[['pre_Price','Train_Number_All','isNormCabin','isWeekend','Customer_Cat','Gap','isdate_1819']])
df['error'] = df['Quantity'] - predicted_quantity
sargan_test_model = ols('error ~ Train_Number_All + isNormCabin + isWeekend + Customer_Cat + Gap + isdate_1819 + isReturn + isOneway', df).fit()
sargan_test_model.summary2()
```

```python
test_stat = sargan_test_model.rsquared * len(df)
p_value = 1 - scipy.stats.chi2.cdf(test_stat,1)
print('test_stat: ' + str(test_stat) + '\np_value: ' + str(p_value))
```
